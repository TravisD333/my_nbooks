```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
av = pd.read_csv('avocado.csv')
```

```python
av.head()
```

```python
av.info()
```

```python
av['Date'] = pd.to_datetime(av['Date'])
```

```python
av.info()
```

```python
av.query('year == "2015" or type == "conventional" or region == "Albany"')[['Date','Total Bags','Small Bags']].head()
```

```python
av_melted = pd.melt(av, id_vars='Date',
                    value_vars=['Total Bags','Small Bags'],
                    var_name='Bags', value_name='Count')
```

```python
av_melted
```

```python
sns.relplot(data=av_melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
av['Sales Volume'] = pd.qcut(av['Total Volume'], q=4,
        labels=['poor','modest','good','excellent'])
```

```python
av.head()
```

```python
sns.countplot(data=av, x='year', hue='Sales Volume')
```

```python
av_melted = pd.melt(av, id_vars='year',
                    value_vars=['Small Bags','Large Bags','XLarge Bags'],
                    var_name='Bags', value_name='Count')
```

```python
av_melted
```

```python
sns.relplot(data=av_melted, kind='line', x='year', y='Count', hue='Bags')
```
